```python
import pandas as pd
```

```python
students_df = pd.read_csv("students_complete.csv")
students_df.head(10)
```

```python
student_names = students_df["student_name"].tolist()
```

```python
# Split the student name and determine the length of the split name to find wrong prefixes attached to names
student_names_to_fix = []

for name in student_names:
    if len(name.split()) >= 3:
        student_names_to_fix.append(name)
        
print(len(student_names_to_fix))
```

```python
prefixes = []

for name in student_names_to_fix:
    if len(name.split()[0]) <= 4:
        prefixes.append(name.split()[0])

unique_prefixes = set(prefixes)
print(unique_prefixes)
```

```python
# Add the suffixes less than or equal to 3 to a new list.
suffixes = []
for name in student_names_to_fix:
    if len(name.split()[-1]) <= 3:
        suffixes.append(name.split()[-1])

unique_suffixes = set(suffixes)
print(unique_suffixes)
```

```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]


for word in prefixes_suffixes:
    students_df["student_name"] = students_df["student_name"].str.replace(word,"")
    
students_df.head(10)
```

```python
# Put the cleaned students' names in another list.
student_names = students_df["student_name"].tolist()
```

```python
# Create a new list and use it for the for loop to iterate through the list.
students_fixed = []

# Use an if statement to check the length of the name.

# If the name is greater than or equal to 3, add the name to the list.

for name in student_names:
    if len(name.split()) >= 3:
        students_fixed.append(name)

# Get the length of the students' names that are greater than or equal to 3.
len(students_fixed)
```
